```python
import pandas as pd

data = {
    "chr" : [],
    "start_pos" : [],
    "end_pos" : [],
    "mod" : [],
    "score" : [],
    "strand" : []
}

df_can = pd.DataFrame(data) #for chunks extraction from the canonical reads
df_un_mod = pd.DataFrame(data) #for chunks extraction from the modified reads of non-modified G
df_mod = pd.DataFrame(data) #for chunks extraction from the modified reads of the modified G's that are located in the middle
df_can_A = pd.DataFrame(data) #for chunks extraction from the canonical reads of A's
df_un_mod_A = pd.DataFrame(data) #for chunks extraction from the modified reads of A's

#the reference file of the canonical reads
fasta = open("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\ref_can.fa", "rt")

seq = ""

p = 0 #can index
q = 0 #can A index

for line in fasta:
    if line[0] == ">":
        seq = line[1:-1]
        print(seq)
    else:
        for i in range(0,len(line)):
            if line[i] == "G":
                df_can.loc[p] = [seq, i, i+1, "pos", "0", "+"]
                p=p+1
            
            if line[i] == "A":
                df_can_A.loc[q] = [seq, i, i+1, "pos", "0", "+"]
                q=q+1

fasta.close()

#choosing the G's that are really within the random area, 4 nucleotides inside - meaning that these G's also have random context.
sub_can = df_can[df_can["start_pos"] > 12]
sub_can = df_can[df_can["start_pos"] < 65]
sub_can_A = df_can_A[df_can_A["start_pos"] > 12]
sub_can_A = df_can_A[df_can_A["start_pos"] < 65]



#the reference file of the modified reads
fasta = open("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\ref_G_cA.fa", "rt")

seq = ""

p = 0 #mod index
q = 0 #un-mod-G index
r = 0 #un-mod-A index

for line in fasta:
    if line[0] == ">":
        seq = line[1:-1]
        print(seq)
    else:
        for i in range(0,len(line)):
            if line[i] == "G":
                if len(line) == 21: #the length of the reads from the can files
                    df_mod.loc[p] = [seq, i, i+1, "pos", "0", "+"]
                    p=p+1
                else:
                    df_un_mod.loc[q] = [seq, i, i+1, "pos", "0", "+"]
                    q=q+1
            
            if line[i] == "A":
                df_un_mod_A.loc[r] = [seq, i, i+1, "pos", "0", "+"]
                r=r+1

fasta.close()

sub_un_mod = df_un_mod[df_un_mod["start_pos"] > 12]
sub_un_mod = df_un_mod[df_un_mod["start_pos"] < 29]

sub_mod = df_mod[df_mod["start_pos"] > 12]
sub_mod = df_mod[df_mod["start_pos"] < 29]

sub_un_mod_A = df_un_mod_A[df_un_mod_A["start_pos"] > 12]
sub_un_mod_A = df_un_mod_A[df_un_mod_A["start_pos"] < 29]


sub_can.to_csv("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\focus_can_G.bed", sep='\t', index=False, header=False)
sub_un_mod.to_csv("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\focus_un_mod_G_cA.bed", sep='\t', index=False, header=False)
sub_mod.to_csv("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\focus_mod_cA.bed", sep='\t', index=False, header=False)
sub_can_A.to_csv("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\focus_can_A.bed", sep='\t', index=False, header=False)
sub_un_mod_A.to_csv("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\focus_un_mod_A_cA.bed", sep='\t', index=False, header=False)
```
