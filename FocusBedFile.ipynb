```python
import pandas as pd

data = {
    "chr" : [],
    "start_pos" : [],
    "end_pos" : [],
    "mod" : [],
    "score" : [],
    "strand" : []
}

df_can = pd.DataFrame(data) #for chunks extraction from the canonical reads
df_un_mod = pd.DataFrame(data) #for chunks extraction from the modified reads of non-modified G
df_mod = pd.DataFrame(data) #for chunks extraction from the modified reads of the modified G's that are located in the middle

df_can_A = pd.DataFrame(data) #for chunks extraction from the canonical reads of A's
df_un_mod_A = pd.DataFrame(data) #for chunks extraction from the modified reads of A's



print(df_can)
print(df_un_mod)
print(df_mod)
print(df_can_A)
print(df_un_mod_A)
```

```python
#the reference file in which all the reads can be found
fasta = open("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\alternative-training\\ref.fa", "rt")

seq = ""

p = 0 #can index
q = 0 #un-mod index
r = 0 #mod index
s = 0 #can A index
t = 0 #un-mod A index

for line in fasta:
    if line[0] == ">":
        seq = line[1:-1]
        print(seq)
    else:
        for i in range(0,len(line)):
            if line[i] == "G":
                if len(line) == 80: #the length of the reads from the can files
                    df_can.loc[p] = [seq, i, i+1, "pos", "0", "+"]
                    p = p+1
                elif i==21:
                    df_mod.loc[r] = [seq, i, i+1, "pos", "0", "+"]
                    r = r+1
                else:
                    df_un_mod.loc[q] = [seq, i, i+1, "pos", "0", "+"]
                    q = q+1
            
            if line[i] == "A":
                if len(line) == 80: #the length of the reads from the can files
                    df_can_A.loc[s] = [seq, i, i+1, "pos", "0", "+"]
                    s = s+1
                else:
                    df_un_mod_A.loc[t] = [seq, i, i+1, "pos", "0", "+"]
                    t = t+1

fasta.close()
```

```python
sub_can = df_can[df_can["start_pos"] > 12]
sub_can = df_can[df_can["start_pos"] < 65]

sub_can_A = df_can_A[df_can_A["start_pos"] > 12]
sub_can_A = df_can_A[df_can_A["start_pos"] < 65]

sub_un_mod = df_un_mod[df_un_mod["start_pos"] > 12]
sub_un_mod = df_un_mod[df_un_mod["start_pos"] < 29]

sub_mod = df_mod[df_mod["start_pos"] > 12]
sub_mod = df_mod[df_mod["start_pos"] < 29]

sub_un_mod_A = df_un_mod_A[df_un_mod_A["start_pos"] > 12]
sub_un_mod_A = df_un_mod_A[df_un_mod_A["start_pos"] < 29]
```
