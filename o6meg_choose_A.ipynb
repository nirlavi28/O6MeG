```python
import re

#working on mod.fastq:
A_w = open("A_w.fastq", "x") #reads where the O6MeG were read as A
A_p = open("A_p.fastq", "x") #only the part of each read from above that is identical to the sequence we are looking for 
G_w = open("G_w.fastq", "x")
G_p = open("G_p.fastq", "x")
AG_p = open("AG_p.fastq", "x") #looking at all the reads that match

fastq = open("C:\\Users\\NirLavi\\Desktop\\PhD\\Lab\\IL LAB\\O6Meg\\data\\exp2_randomers\\fastqc\\mod.fastq", "rt")

curr_id = ""
curr_read = ""

matches_A = 0
matches_G = 0

for line in fastq:
    if line[0] == '@': #id line
        curr_id = line
    elif line[0] in ['A','T','G','C']: #read line
        curr_read = line
        matches_A = re.finditer(r'ATGAGTCTA............A............ACGCACACG', line)
        matches_G = re.finditer(r'ATGAGTCTA............G............ACGCACACG', line)
    elif line[0] != '+': #the quality line, and the last line regarding a certain read
        curr_quality = line
        
        first = True
        for match in matches_A:
            if first:
                A_w.write(curr_id)
                A_w.write(curr_read)
                A_w.write('+\n')
                A_w.write(curr_quality)
                first = False
            
            A_p.write(curr_id)
            A_p.write(curr_read[match.start():match.end()] + '\n')
            A_p.write('+\n')
            A_p.write(curr_quality[match.start():match.end()] + '\n')
            #write the read also in AG_p:
            AG_p.write(curr_id)
            AG_p.write(curr_read[match.start():match.end()] + '\n')
            AG_p.write('+\n')
            AG_p.write(curr_quality[match.start():match.end()] + '\n')

        first = True
        for match in matches_G:
            if first:
                G_w.write(curr_id)
                G_w.write(curr_read)
                G_w.write('+\n')
                G_w.write(curr_quality)
                first = False
            
            G_p.write(curr_id)
            G_p.write(curr_read[match.start():match.end()] + '\n')
            G_p.write('+\n')
            G_p.write(curr_quality[match.start():match.end()] + '\n')
            #write the read also in AG_p:
            AG_p.write(curr_id)
            AG_p.write(curr_read[match.start():match.end()] + '\n')
            AG_p.write('+\n')
            AG_p.write(curr_quality[match.start():match.end()] + '\n')
                        
fastq.close()
A_w.close()
A_p.close()
G_w.close()
G_p.close()
AG_p.close()
```
